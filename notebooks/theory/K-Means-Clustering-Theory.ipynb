# K Means Clustering

K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. In k means clustering, we have the specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster and finds the centroid of each cluster. Then, the algorithm iterates through two steps: Reassign data points to the cluster whose centroid is closest. Calculate the new centroid of each cluster. These two steps are repeated till the within cluster variation cannot be reduced any further. The within cluster variation is calculated as the sum of the euclidean distance between the data points and their respective cluster centroids.

## Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Create some data

```python
from sklearn.datasets import make_blobs
```

We can create a data set using the `make_blobs` functions, defining the number of columns with the argument` n_features`, number of rows with `n_samples`, and number of classes with` centers`.

```python
data = make_blobs(n_samples=200, n_features=2, 
                     centers=4, cluster_std=1.8, random_state=101)
```

## Visualize Data

```python
sns.set_style('darkgrid')
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

## Creating the Clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

We can identify the labels using this method

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

We can notice that in the center of the three clusters in the upper right region there is a relative noise, causing difficulty in the accurate identification of K Means. However, it is possible to observe that the algorithm did a good job in the classification. 

It's important to note that in this example we use the labels from the original data set to predict the evaluation of the algorithm. In real-world problems we, probably, don't have the labels!
